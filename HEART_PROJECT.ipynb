```python
#feature engineering

import sklearn.preprocessing as skl
from sklearn.model_selection import KFold

X=data1[['age','sex','cp','trestbps','chol','restecg','thalach','fbs','thal','slope']]
y=data1['target']
k = 10  
k_fold = KFold(n_splits = k, random_state = None)  
sc=skl.StandardScaler()
da=sc.fit_transform(X,y)
```

```python
sns.heatmap(data1.corr(),annot=True)
```

```python
#checking generalisation of the data with various classifiers

#Random FOREST
from sklearn.ensemble import RandomForestClassifier
Random_model = RandomForestClassifier()
Random_model.fit(X,y)
print("Random Forest Classifier accuracy: ", (Random_model.score(X,y))*100)

from sklearn.tree import DecisionTreeClassifier
decision_model = DecisionTreeClassifier()
decision_model.fit(X,y)
print("Decision Tree Classifier accuracy: ", (decision_model.score(X,y))*100)

from xgboost import XGBClassifier
xgb_model = XGBClassifier(n_estimators=390, learning_rate=0.1)
xgb_model.fit(X,y)
print("XGBoost Classifier accuracy: ", (xgb_model.score(X,y))*100)

from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
knn_model.fit(X,y)
print("KNN Classifier accuracy", (knn_model.score(X,y))*100)

from sklearn.naive_bayes import GaussianNB
NB_model = GaussianNB()
NB_model.fit(X,y)
print("Gaussion Navie Bayes accuracy: " ,(NB_model.score(X,y))*100)

from sklearn.svm import SVC
svm_model=SVC()
svm_model.fit(X,y)
print("SVM accuracy: ", (svm_model.score(X,y))*100)

from sklearn.ensemble import AdaBoostClassifier
adb_model = AdaBoostClassifier(n_estimators=390, learning_rate=0.1)
adb_model.fit(X,y)
print("ADBoost Classifier accuracy: ", (adb_model.score(X,y))*100)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=150)
y_pred=adb_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("accuracy of the model USING ADABOOST: ", accuracy_score(y_test, adb_model.predict(X_test))*100)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average='binary')
print("precision: ", precision*100)
```

```python
from sklearn.metrics import recall_score
recall = precision_score(y_test, y_pred, average='binary')
print("recall: ", recall*100)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("r2: ", r2*100)
```
